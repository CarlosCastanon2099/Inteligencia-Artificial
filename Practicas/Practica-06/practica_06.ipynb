- ### <span style="color:#17FC3F">**Pregunta 1**</span>

    #### üìù **Investiga y redacta un breve resumen sobre qu√© son las redes bayesianas y para qu√© se utilizan.**

    Una **red bayesiana** es un gr√°fico en el que los nodos representan entidades como mol√©culas o genes. Los nodos que interact√∫an est√°n conectados por bordes en la direcci√≥n de influencia; la arista $A \to B$ implica que $A$ (el padre) tiene un efecto sobre $B$ (el hijo). En general, una red bayesiana es un gr√°fico ac√≠clico dirigido; es decir, no se permiten ciclos. Es importante destacar que cada nodo tiene probabilidades adjuntas que definen la posibilidad de encontrar el nodo en un estado determinado. Las probabilidades condicionales se utilizan si el estado de un nodo depende del estado de otro. Estas dependencias se propagan a trav√©s de la red e influyen en las probabilidades de otros nodos, que se actualizan a medida que hay nueva informaci√≥n disponible sobre los nodos. Por tanto, las redes bayesianas tambi√©n se denominan **modelos causales probabil√≠sticos**.
        
    Las redes bayesianas se utilizan en una variedad de campos para modelar la incertidumbre y tomar decisiones bajo esta. Algunas de las aplicaciones comunes incluyen: diagn√≥stico m√©dico, an√°lisis de riesgos, reconocimiento de patrones. ,odelado de sistemas expertos o sistemas de recomendaci√≥n.
        
    Un ejemplo de esto podr√≠a ser un analista de un banco que quiere poder predecir clientes o clientes potenciales, con posibilidades de impago de sus cr√©ditos. Para ellos, puede utilizar un modelo de red bayesiana para identificar las caracter√≠sticas de los clientes con m√°s posibilidades de impago y generar varios tipos diferentes de modelo para establecer el mejor m√©todo para predecir los clientes con m√°s posibilidades de impago.

- ### <span style="color:#17FC3F">**Pregunta 2**</span>

    #### ‚ûó **Explica la diferencia entre probabilidad condicional e independencia condicional, incluyendo ejemplos claros de cada uno.**

    ##### <span style="color:#0082F5">**Probabilidad Condicional**</span>

    La **probabilidad condicional** se refiere a la probabilidad de que ocurra un evento dado que otro evento ya ha ocurrido. Se denota como $P(A | B)$ que significa la probabilidad de que ocurra el evento $A$ dado que el evento $B$. 

    Por lo tanto, podemos decir que la **f√≥rmula para calcular la probabilidad condicional** es: $P(A | B) = \frac{P(A \cap B)}{P(B)}$

    > **Ejemplo:**
    >
    > - Calcular la probabilidad de enfermedad dado un resultado de prueba:
    >
    >   Supongamos que estamos interesados en la probabilidad de que una persona est√© enferma $(E)$ dado que el resultado de una prueba $(T)$ es positivo. La probabilidad condicional en este caso ser√≠a $P(E|T)$, es decir, la probabilidad de que la persona est√© enferma dado que la prueba dio positivo. Esta probabilidad puede ser diferente de la probabilidad de estar enfermo sin tener en cuenta el resultado de la prueba. 
    >    
    >   Entonces, tendremos la siguiente f√≥rmula: $P(E|T) = \frac{P(E \cap T)}{P(T)}$

    ##### <span style="color:#F5CB0F">****Independencia Condicional****</span>

    La **independencia condicional** ocurre cuando la ocurrencia de un evento no afecta la probabilidad de otro evento. Formalmente, dos eventos $A$ y $B$ son independientes condicionalmente si la probabilidad de que ocurra $A$ dado que $B$ ya ha ocurrido es la misma que la probabilidad de que ocurra $A$ en general. 

    Por lo tanto, podemos decir que la **la expresi√≥n matem√°tica de la independencia condicional** es: $P(A|B) = P(A)$

    > **Ejemplo:**
    >
    > - Calcular la independencia Condicional de un lanzamiento de dos monedas:
    >
    >    Supongamos que estamos lanzando dos monedas. Los eventos $A$ y $B$ podr√≠an ser "la primera moneda da cara" y "la segunda moneda da cara", respectivamente. Si las dos monedas est√°n justas (es decir, la probabilidad de obtener cara en cada lanzamiento es $\frac{1}{2}$ entonces la ocurrencia de que la primera moneda muestre cara no afecta la probabilidad de que la segunda moneda muestre cara. Por lo tanto, en este caso, los eventos $A$ y $B$ son independientes condicionalmente, 
    >
    >    Entonces, podemos denotar a este evento como $P(A|B) = P(A) = \frac{1}{2}$

- ### <span style="color:#17FC3F">**Pregunta 3**</span>

    #### üìê **Describe c√≥mo se representa matem√°ticamente una red bayesiana y el significado de los nodos y aristas en cada representaci√≥n.**

    Una **red bayesiana** es un modelo matem√°tico para representar relaciones causales entre variables aleatorias mediante el uso de probabilidades condicionales. En el contexto de una red bayesiana, asumimos que existe un gr√°fico ac√≠clico dirigido, denotado por $G$, como una relaci√≥n entre variables aleatorias. En la estimaci√≥n de redes gen√©ticas basada en redes bayesianas, un gen se considera una variable aleatoria y se muestra como un nodo. Sea $X_i  $ una variable aleatoria discreta que toma un valor de $\{u_1,‚Ä¶ , u_m\}$, si hay una arista dirigida $e_{ij}$ de $X_i$ a $X_j$, llamamos a $X_i$ padre de $X_j$, por lo que podemos notar que las aristas representan una relaci√≥n causal.

    Adem√°s, definimos $Pa(X_i) \subseteq \{X1, ‚Ä¶, X_p\}$ como el conjunto de padres de $X_i$ en $G$. En $G$, la variable aleatoria $X_i$ s√≥lo depende de sus padres directos $Pa(X_i)$ y es independiente de otros variables (es decir, esto ofrece la propiedad de Markov de primer orden a la relaci√≥n entre variables descrita por $G$). Usando G y su propiedad de Markov, la probabilidad conjunta de todas las variables aleatorias se puede descomponer como el producto de probabilidades condicionales:

    $P(X_i, ..., X_p) = \prod_{j = 1}^{p} P(X_j \mid Pa (X_j))$

    Como $X_j$ es una variable discreta, las probabilidades $\vartheta_{jkl} = P(X_j = u_k \mid Pa(X_j) = u_{jl}) (j = 1, ‚Ä¶, p; k = 1, ‚Ä¶, m; l = 1, ‚Ä¶, m \mid Pa (X_i)|)$ son par√°metros, donde $u_{jl}$ es la l-√©sima entrada de la tabla de estados de los padres de $X_j$ y $|Pa(X_j)|$ es el n√∫mero de padres de $X_j$.

    Por lo tanto, las probabilidades condicionales $P(X_j \mid Pa(X_j))$ describen las relaciones padre-hijo y pueden verse como una extensi√≥n de los modelos deterministas, como las redes booleanas.

### üëæ **Implementaci√≥n de Redes Bayesianas con pgmpy**

Importamos las bibliotecas necesarias y establecemos las conexiones de los nodos:
- **H**: Horas de trabajo
- **B**: Balance trabajo-vida
- **S**: Satisfacci√≥n laboral

```python
#usar !pip install pgmpy
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD

# Se define la red bayesiana
modelo = BayesianNetwork([('B', 'S'), ('H', 'B')])

```

Creamos las tablas de probabilidad condicional para cada uno de los nodos descritos anteriormente. Y las imprimos para poder visualizar de una manera gr√°fica y legible su estructura.

```python
cpd_h = TabularCPD(variable='H',
                   variable_card=3, values=[[0.24], [0.72], [0.04]], #Probabilidades
                   state_names={'H': ['Largas', 'Moderadas', 'Cortas']})

cpd_b = TabularCPD(variable='B', variable_card=2,
                      values=[[0.13, 0.35, 0.72], #Probabilidades
                              [0.87, 0.65, 0.28]],
                      evidence=['H'],
                      evidence_card=[3],
                      state_names={'B': ['Equilibrado', 'No equilibrado'],
                                   'H': ['Largas', 'Moderadas', 'Cortas']})

cpd_s = TabularCPD(variable='S', variable_card=3,
                      values=[[0.37, 0.02], #Probabilidades
                              [0.49, 0.29],
                              [0.14, 0.69]],
                      evidence=['B'],
                      evidence_card=[2],
                      state_names={'S': ['Satisfecho', 'Neutral', 'Insatisfecho'],
                                   'B': ['Equilibrado', 'No equilibrado']})

print("Tabla de probabilidad condicional para las horas de trabajo")
print(cpd_h)
print()
print("Tabla de probabilidad condicional para el balance trabajo-vida")
print(cpd_b)
print()
print("Tabla de probabilidad condicional para la satsifacci√≥n laboral")
print(cpd_s)
print()
```

Posteriormente se a√±aden las tablas de probabilidad condicional a la red bayesiana y se verifica que el modelo sea v√°lido. Esto es cierto, si y s√≥lo si, la suma de las probabilidades de cada tabla sea igual a 1.

```python
# Agrega tablas de probabilidad condicional a la red bayesiana
modelo.add_cpds(cpd_h)
modelo.add_cpds(cpd_b)
modelo.add_cpds(cpd_s)

# verifica que la suma de las tablas sea 1
if modelo.check_model:
    print('La red bayesiana es v√°lida')
```

Ahora creamos una instancia de la clase VariableElimination que nos permite realizar predicciones con la red bayesiana construida anteriormente y verificamos las probabilidades de satsisfacci√≥n dado que las horas de trabajo sean moderadas y el balance trabajo-vida es equilibrado.

```python
from pgmpy.inference import VariableElimination

inferencia = VariableElimination(modelo)
# Obtiene probabilidades de la variable S cuando tiene horas de trabajo moderadas y un balance trabajo-vida equilibrado
probabilidades_s = inferencia.query(variables = ['S'], evidence = {'B': 'Equilibrado', 'H': 'Moderadas'})
print()
print("Tabla de probabilidades de satisfacci√≥n laboral con condiciones dadas")
print(probabilidades_s)
```

### ‚öôÔ∏è **Ajuste de probabilidades**

Tomemos en cuenta la siguiente situaci√≥n dentro de nuestra red bayesiana
- El balance vida-trabajo se considera m√°s balanceado cuando se trabajan horas cortas.

Y lo cambiaremos por

- El balance vida-trabajo se considera m√°s balanceado cuando se trabajan horas moderadas.

Implementemos el cambio dentro del c√≥digo.

```python
# Modificamos dicha probabilidad
cpd_s = TabularCPD(variable='S', variable_card=3,
                      values=[[0.67, 0.02], #Probabilidades
                              [0.26, 0.69],
                              [0.07, 0.29]],
                      evidence=['B'],
                      evidence_card=[2],
                      state_names={'S': ['Satisfecho', 'Neutral', 'Insatisfecho'],
                                   'B': ['Equilibrado', 'No equilibrado']})

# Reestablecemos nuestro modelo
modelo = BayesianNetwork([('B', 'S'), ('H', 'B')])

#Agregamos las tablas de probabilidad condicional
modelo.add_cpds(cpd_h)
modelo.add_cpds(cpd_b)
modelo.add_cpds(cpd_s)

# verifica que la suma de las tablas sea 1
if modelo.check_model:
    print('La red bayesiana es v√°lida')

inferencia = VariableElimination(modelo)
# Obtiene probabilidades de la variable S cuando tiene horas de trabajo moderadas y un balance trabajo-vida no equilibrado
probabilidades_s = inferencia.query(variables = ['S'], evidence = {'B': 'Equilibrado', 'H': 'Moderadas'})
print()
print("Tabla de probabilidades de satisfacci√≥n laboral con condiciones dadas")
print(probabilidades_s)
```
