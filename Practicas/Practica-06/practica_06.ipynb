- ### <span style="color:#17FC3F">**Pregunta 1**</span>

    #### 📝 **Investiga y redacta un breve resumen sobre qué son las redes bayesianas y para qué se utilizan.**

    Una **red bayesiana** es un gráfico en el que los nodos representan entidades como moléculas o genes. Los nodos que interactúan están conectados por bordes en la dirección de influencia; la arista $A \to B$ implica que $A$ (el padre) tiene un efecto sobre $B$ (el hijo). En general, una red bayesiana es un gráfico acíclico dirigido; es decir, no se permiten ciclos. Es importante destacar que cada nodo tiene probabilidades adjuntas que definen la posibilidad de encontrar el nodo en un estado determinado. Las probabilidades condicionales se utilizan si el estado de un nodo depende del estado de otro. Estas dependencias se propagan a través de la red e influyen en las probabilidades de otros nodos, que se actualizan a medida que hay nueva información disponible sobre los nodos. Por tanto, las redes bayesianas también se denominan **modelos causales probabilísticos**.
        
    Las redes bayesianas se utilizan en una variedad de campos para modelar la incertidumbre y tomar decisiones bajo esta. Algunas de las aplicaciones comunes incluyen: diagnóstico médico, análisis de riesgos, reconocimiento de patrones. ,odelado de sistemas expertos o sistemas de recomendación.
        
    Un ejemplo de esto podría ser un analista de un banco que quiere poder predecir clientes o clientes potenciales, con posibilidades de impago de sus créditos. Para ellos, puede utilizar un modelo de red bayesiana para identificar las características de los clientes con más posibilidades de impago y generar varios tipos diferentes de modelo para establecer el mejor método para predecir los clientes con más posibilidades de impago.

- ### <span style="color:#17FC3F">**Pregunta 2**</span>

    #### ➗ **Explica la diferencia entre probabilidad condicional e independencia condicional, incluyendo ejemplos claros de cada uno.**

    ##### <span style="color:#0082F5">**Probabilidad Condicional**</span>

    La **probabilidad condicional** se refiere a la probabilidad de que ocurra un evento dado que otro evento ya ha ocurrido. Se denota como $P(A | B)$ que significa la probabilidad de que ocurra el evento $A$ dado que el evento $B$. 

    Por lo tanto, podemos decir que la **fórmula para calcular la probabilidad condicional** es: $P(A | B) = \frac{P(A \cap B)}{P(B)}$

    > **Ejemplo:**
    >
    > - Calcular la probabilidad de enfermedad dado un resultado de prueba:
    >
    >   Supongamos que estamos interesados en la probabilidad de que una persona esté enferma $(E)$ dado que el resultado de una prueba $(T)$ es positivo. La probabilidad condicional en este caso sería $P(E|T)$, es decir, la probabilidad de que la persona esté enferma dado que la prueba dio positivo. Esta probabilidad puede ser diferente de la probabilidad de estar enfermo sin tener en cuenta el resultado de la prueba. 
    >    
    >   Entonces, tendremos la siguiente fórmula: $P(E|T) = \frac{P(E \cap T)}{P(T)}$

    ##### <span style="color:#F5CB0F">****Independencia Condicional****</span>

    La **independencia condicional** ocurre cuando la ocurrencia de un evento no afecta la probabilidad de otro evento. Formalmente, dos eventos $A$ y $B$ son independientes condicionalmente si la probabilidad de que ocurra $A$ dado que $B$ ya ha ocurrido es la misma que la probabilidad de que ocurra $A$ en general. 

    Por lo tanto, podemos decir que la **la expresión matemática de la independencia condicional** es: $P(A|B) = P(A)$

    > **Ejemplo:**
    >
    > - Calcular la independencia Condicional de un lanzamiento de dos monedas:
    >
    >    Supongamos que estamos lanzando dos monedas. Los eventos $A$ y $B$ podrían ser "la primera moneda da cara" y "la segunda moneda da cara", respectivamente. Si las dos monedas están justas (es decir, la probabilidad de obtener cara en cada lanzamiento es $\frac{1}{2}$ entonces la ocurrencia de que la primera moneda muestre cara no afecta la probabilidad de que la segunda moneda muestre cara. Por lo tanto, en este caso, los eventos $A$ y $B$ son independientes condicionalmente, 
    >
    >    Entonces, podemos denotar a este evento como $P(A|B) = P(A) = \frac{1}{2}$

- ### <span style="color:#17FC3F">**Pregunta 3**</span>

    #### 📐 **Describe cómo se representa matemáticamente una red bayesiana y el significado de los nodos y aristas en cada representación.**

    Una **red bayesiana** es un modelo matemático para representar relaciones causales entre variables aleatorias mediante el uso de probabilidades condicionales. En el contexto de una red bayesiana, asumimos que existe un gráfico acíclico dirigido, denotado por $G$, como una relación entre variables aleatorias. En la estimación de redes genéticas basada en redes bayesianas, un gen se considera una variable aleatoria y se muestra como un nodo. Sea $X_i  $ una variable aleatoria discreta que toma un valor de $\{u_1,… , u_m\}$, si hay una arista dirigida $e_{ij}$ de $X_i$ a $X_j$, llamamos a $X_i$ padre de $X_j$, por lo que podemos notar que las aristas representan una relación causal.

    Además, definimos $Pa(X_i) \subseteq \{X1, …, X_p\}$ como el conjunto de padres de $X_i$ en $G$. En $G$, la variable aleatoria $X_i$ sólo depende de sus padres directos $Pa(X_i)$ y es independiente de otros variables (es decir, esto ofrece la propiedad de Markov de primer orden a la relación entre variables descrita por $G$). Usando G y su propiedad de Markov, la probabilidad conjunta de todas las variables aleatorias se puede descomponer como el producto de probabilidades condicionales:

    $P(X_i, ..., X_p) = \prod_{j = 1}^{p} P(X_j \mid Pa (X_j))$

    Como $X_j$ es una variable discreta, las probabilidades $\vartheta_{jkl} = P(X_j = u_k \mid Pa(X_j) = u_{jl}) (j = 1, …, p; k = 1, …, m; l = 1, …, m \mid Pa (X_i)|)$ son parámetros, donde $u_{jl}$ es la l-ésima entrada de la tabla de estados de los padres de $X_j$ y $|Pa(X_j)|$ es el número de padres de $X_j$.

    Por lo tanto, las probabilidades condicionales $P(X_j \mid Pa(X_j))$ describen las relaciones padre-hijo y pueden verse como una extensión de los modelos deterministas, como las redes booleanas.
