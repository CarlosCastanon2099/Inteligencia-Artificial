### üìè **Fundamentos de Redes Bayesianas**

##### 1. üìù Investiga y redacta un breve resumen (m√°ximo una cuartilla) sobre qu√© son las redes bayesianas y para qu√© se utilizan.

Una red bayesiana es un gr√°fico en el que los nodos representan entidades como mol√©culas o genes. Los nodos que interact√∫an est√°n conectados por bordes en la direcci√≥n de influencia; la arista A‚ÜíB implica que A (el padre) tiene un efecto sobre B (el hijo). En general, una red bayesiana es un gr√°fico ac√≠clico dirigido; no se permiten ciclos. Es importante destacar que cada nodo tiene probabilidades adjuntas que definen la posibilidad de encontrar el nodo en un estado determinado. Las probabilidades condicionales se utilizan si el estado de un nodo depende del estado de otro. Estas dependencias se propagan a trav√©s de la red e influyen en las probabilidades de otros nodos, que se actualizan a medida que hay nueva informaci√≥n disponible sobre los nodos. Por tanto, las redes bayesianas tambi√©n se denominan modelos causales probabil√≠sticos.

Los nodos con variables continuas se parametrizan mediante funciones de probabilidad y los de variables discretas mediante tablas de probabilidad. Por ejemplo, considere la red simple de dos nodos A‚ÜíB donde A y B son variables binarias con dos estados (N o Y). La tabla en el nodo A contendr√≠a la probabilidad marginal P(A = Y). Para simplificar, usaremos "A" para significar A = Y y "a" para significar A = N, de modo que P(A = Y) y P(A = N) se puedan escribir m√°s brevemente como P(A) y P(a), respectivamente. Por complementariedad, P(a) = 1 ‚Äì P(A). En el nodo B tendr√≠amos las probabilidades condicionales P(B|A) y P(Bla) que definen c√≥mo el estado de B depende del estado de A. La tabla de probabilidad condicional (CPT) se puede completar usando complementariedad: P(b |A) = 1 ‚Äì P(B|A) y P(b|a) = 1 ‚Äì P(B|a). Por lo tanto, la tabla marginal para A enumera todos los estados posibles para A, y el CPT enumera todas las combinaciones de estados posibles de A y B. Una vez que se construye la red y se especifican las probabilidades, se utiliza el teorema de Bayes para propagar la probabilidad a trav√©s del modelo.

##### 2. ‚ûó Explica la diferencia entre probabilidad condicional e independencia condicional, incliyendo ejemplos claros de cada uno.

- **Probabilidad Condicional**

    La probabilidad condicional se refiere a la probabilidad de que ocurra un evento dado que otro evento ya ha ocurrido. Se denota como P(A|B)  que significa la probabilidad de que ocurra el evento A dado que el evento B. 

    Formula de probabilidad condicional:
    $P(A|B) = \frac{P(A \cap B)}{P(B)} $

    - **Ejemplo 1:**

        Probabilidad de Enfermedad dado Resultado de Prueba:

        Supongamos que estamos interesados en la probabilidad de que una persona est√© enferma (E) dado que el resultado de una prueba (T) es positivo. La probabilidad condicional en este caso ser√≠a P(E|T), es decir, la probabilidad de que la persona est√© enferma dado que la prueba dio positivo. Esta probabilidad puede ser diferente de la probabilidad de estar enfermo sin tener en cuenta el resultado de la prueba.

        $P(E|T) = P(T|E)*P(E) / P(T)$


----

- **Independencia Condicional**

    La independencia condicional ocurre cuando la ocurrencia de un evento no afecta la probabilidad de otro evento. Formalmente, dos eventos
    A y B son independientes condicionalmente si la probabilidad de que ocurra A dado que B ya ha ocurrido es la misma que la probabilidad de que ocurra A en general. En t√©rminos matem√°ticos, esto se¬†expresa¬†como:

    Formula de independencia condicional:
    $P(A|B) = P(A)$

    - **Ejemplo 1:**

        Independencia Condicional lanzando dos monedas:

        Supongamos que estamos lanzando dos monedas. Los eventos A y B podr√≠an ser "la primera moneda muestra cara" y "la segunda moneda muestra cara", respectivamente. Si las dos monedas est√°n justas (es decir, la probabilidad de obtener cara en cada lanzamiento es $\frac{1}{2}$ ) entonces la ocurrencia de que la primera moneda muestre cara no afecta la probabilidad de que la segunda moneda muestre cara. Por lo tanto, en este caso, los eventos A y son independientes condicionalmente, y

        $P(A|B) = P(A) = \frac{1}{2}$

Una red bayesiana es un modelo matem√°tico para representar relaciones causales entre variables aleatorias mediante el uso de probabilidades condicionales. Sea Xi (i = 1,‚Ä¶, p) una variable aleatoria discreta que toma un valor de {u1,‚Ä¶, um}. Si hay una arista dirigida eij de Xi a Xj, llamamos a Xi padre de Xj.

Adem√°s, definimos Pa(Xi) ‚äÜ {X1, ‚Ä¶, Xp} como el conjunto de padres de Xi en G. En el DAG G, la variable aleatoria Xi s√≥lo depende de sus padres directos Pa(Xi) y es independiente de otros variables (es decir, esto ofrece la propiedad de Markov de primer orden a la relaci√≥n entre variables descrita por G). Usando el DAG G y su propiedad de Markov, la probabilidad conjunta de todas las variables aleatorias se puede descomponer como el producto de probabilidades condicionales:

Como Xj es una variable discreta, las probabilidades Œ∏jkl = P(Xj = uk|Pa(Xj) = ujl) (j = 1, ‚Ä¶, p; k = 1, ‚Ä¶, m; l = 1, ‚Ä¶, m|Pa (Xi)|) son par√°metros, donde ujl es la l-√©sima entrada de la tabla de estados de los padres de Xj y |Pa(Xj)| es el n√∫mero de padres de Xj. Por ejemplo, para |Pa(Xj)| = 2 tenemos uj1 = (u1, u1), uj2 = (u1, u2), y as√≠ sucesivamente. Las probabilidades condicionales P(Xj |Pa(Xj)) describen las relaciones padre-hijo y pueden verse como una extensi√≥n de los modelos deterministas, como las redes booleanas (Somogyi y Sniegoski 1996).
